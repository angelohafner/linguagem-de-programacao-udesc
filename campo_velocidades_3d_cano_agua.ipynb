```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plotly.graph_objects as go
```

```python

nr_pontos = 12*4
theta = np.linspace(0, 2*np.pi, nr_pontos)
raio = np.linspace(0, 1, nr_pontos)
RAIO, THETA = np.meshgrid(raio, theta)
X = RAIO * np.cos(THETA)
Y = RAIO * np.sin(THETA)
```

```python
plt.scatter(X, Y)
plt.scatter(raio*np.cos(theta), raio*np.sin(theta))
```

```python
Z = - ( X**2 + Y**2) + 1
fig = go.Figure(data=[
    go.Surface(x=X, y=Y, z=Z, colorscale="Viridis")
])

fig.update_layout(
    title="Z = -(X² + Y²)+1",
    scene=dict(
        xaxis_title="X",
        yaxis_title="Y",
        zaxis_title="Z"
    ),
    width=800,
    height=600
)

fig.show()
```

```python
nr_pontos = 12*10
theta = np.linspace(0, 2*np.pi, nr_pontos)
raio = np.linspace(0, 1, nr_pontos)
z = np.linspace(0, 0, 1)  # plano z=0

RAIO, THETA, Z = np.meshgrid(raio, theta, z)

X = RAIO*np.cos(THETA)
Y = RAIO*np.sin(THETA)
Z = Z

# Vector components
W = - (X**2 + Y**2) + 1
U = 0 * W
V = 0 * W

# Flatten arrays for plotting
x_flat = X.ravel()
y_flat = Y.ravel()
z_flat = Z.ravel()
u_flat = U.ravel()
v_flat = V.ravel()
w_flat = W.ravel()


# Create 3D quiver using cones
fig = go.Figure(
    data=go.Cone(
        x=x_flat,
        y=y_flat,
        z=z_flat,
        u=u_flat,
        v=v_flat,
        w=w_flat,
        colorscale="Viridis",
        sizemode="absolute",
        sizeref=0.5,  # controla o tamanho das setas
        showscale=True
    )
)

fig.update_layout(
    scene=dict(
        xaxis_title="X",
        yaxis_title="Y",
        zaxis_title="Z",
        aspectratio=dict(x=1, y=1, z=1)
    ),
    title="Quiver 3D com U=0, V=0, W=-(X²+Y²)+1",
    width=800,
    height=600
)

fig.show()
```
