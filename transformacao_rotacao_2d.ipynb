```python
import numpy as np
import matplotlib.pyplot as plt

vetor_entrada = np.array([[1],
                          [1]])

theta = 45 * np.pi/180

a11 =  np.cos(theta)
a12 = -np.sin(theta)
a21 =  np.sin(theta)
a22 =  np.cos(theta)

matriz_rotacao = [[a11, a12],
                  [a21, a22]]

matriz_rotacao = np.array(matriz_rotacao)


vetor_saida = matriz_rotacao @ vetor_entrada
print(vetor_entrada)
print("")
print(vetor_saida)
```

```python
fig, ax = plt.subplots()
ax.quiver(0, 0, vetor_entrada[0], vetor_entrada[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v_in')
ax.quiver(0, 0, vetor_saida[0], vetor_saida[1], angles='xy', scale_units='xy', scale=1, color='green', label='v_out')
ax.set_xlim(-2*vetor_entrada[0], 2*vetor_entrada[0])
ax.set_ylim(-2*vetor_entrada[1], 2*vetor_entrada[1])
ax.set_aspect('equal')
ax.legend()
ax.grid()
```

```python
fig, ax = plt.subplots()
ax.scatter(vetor_entrada[0], vetor_entrada[1], color='blue', label='v_in')
ax.scatter(vetor_saida[0], vetor_saida[1], color='green', label='v_out')
ax.set_xlim(-2*vetor_entrada[0], 2*vetor_entrada[0])
ax.set_ylim(-2*vetor_entrada[1], 2*vetor_entrada[1])
ax.set_aspect('equal')
ax.legend()
ax.grid()
```

```python
pontos_externo = [
    [1, 4, 4, 4.2, 2.5, 0.8, 1, 1],
    [1, 1, 3, 2.7, 5,   2.7, 3, 1]
]

pontos_externo = np.array(pontos_externo)
pontos_externo_girado = matriz_rotacao @ pontos_externo

fig, ax = plt.subplots()
ax.scatter(pontos_externo[0,:], pontos_externo[1,:], color='blue', label='pontos_externo')
ax.plot(pontos_externo[0,:], pontos_externo[1,:], color='blue', label='pontos_externo')
ax.scatter(pontos_externo_girado[0,:], pontos_externo_girado[1,:], color='green', label='pontos_externo_girado')
ax.plot(pontos_externo_girado[0,:], pontos_externo_girado[1,:], color='green', label='pontos_externo_girado')
ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
ax.set_aspect('equal')
ax.grid()
```
