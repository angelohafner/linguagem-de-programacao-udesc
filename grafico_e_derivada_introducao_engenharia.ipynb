# Comparação de funções relacionadas por derivada na Engenharia de Petróleo

## 1) Pressão vs. Derivada da Pressão (Teste de Pressão)
- Função: $$ p(t) $$
- Derivada: $$ \frac{d p}{d \ln t} $$
- Aplicação: identificação de regimes de fluxo, permeabilidade, skin.

---

## 2) Produção Acumulada vs. Taxa de Produção
- Função: $$ N_p(t) $$
- Derivada: $$ q(t) = \frac{dN_p}{dt} $$
- Aplicação: curvas de declínio (Arps), balanço de materiais.

---

## 3) Saturação vs. Derivada da Fração de Fluxo (Buckley–Leverett)
- Função: $$ S_w(x,t) $$
- Derivada: $$ \frac{\partial f_w}{\partial S_w} $$
- Aplicação: velocidade da frente de água, deslocamento imiscível.

---

## 4) Pressão Radial vs. Gradiente de Pressão (Lei de Darcy)
- Função: $$ p(r) $$
- Derivada: $$ \frac{dp}{dr} $$
- Aplicação: cálculo de vazão radial, fluxo no reservatório.

---


```python
import numpy
import matplotlib.pyplot

x = numpy.arange(2, 8*numpy.pi, 1e-3)
y = 0
for numero in range(1, 10, 2):
    y = y + 10 / numero * numpy.sin(numero*x)
```

```python
matplotlib.pyplot.figure(figsize=(10, 4))
matplotlib.pyplot.plot(x, y)
```

```python
delta_x = numpy.diff(x)
len(delta_x)
delta_y = numpy.diff(y)
inclinacao = delta_y / delta_x
```

```python
print(f"comprimento de x = {len(x)} e de Δx = {len(delta_x)}")
```

```python
matplotlib.pyplot.figure(figsize=(10, 4))
matplotlib.pyplot.plot(x[0:len(x)-1], inclinacao, label='$\\frac{d}{dx} f(x)$', ls='--')
matplotlib.pyplot.plot(x, y, label='$f(x)$')
matplotlib.pyplot.grid()
matplotlib.pyplot.legend()
```
