```python
import matplotlib.pyplot as plt
import numpy as np
dx = 0.1
x = np.arange(-2, 2+dx, dx)
y = x**2

fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title("Grfico do Seno")
ax.set_xlabel(r"$x$")
ax.set_ylabel(r"$y=\sin x$")
ax.grid(ls=":")
```

```python
dx = 0.1
x = np.arange(-2, 2+dx, dx)
dy = 0.1
y = np.arange(-2, 2+dy, dy)
X, Y = np.meshgrid(x, y)
fig, ax = plt.subplots()

ax.scatter(X,Y)
ax.scatter(x,y)
```

```python
Z = X**2 + Y**2
fig3, ax3 = plt.subplots(subplot_kw={"projection": "3d"})
ax3.plot_surface(X, Y, Z)
```

```python
fig5, ax5 = plt.subplots()
CS = ax5.contourf(X, Y, Z)
fig5.colorbar(CS, ax=ax5)
```

```python
fig6 = plt.figure()
ax6 = fig6.add_subplot(111, projection='3d')
CS = ax6.contour3D(X, Y, Z)
fig6.colorbar(CS, ax=ax6)
```

```python
fig4, ax4 = plt.subplots()
CS = ax4.contour(X, Y, Z)
ax.clabel(CS, fontsize=10)
```

```python
gradiente_temperatura = np.gradient(Z, dy, dx)
dZ_dy = gradiente_temperatura[0]   # componente em y
dZ_dx = gradiente_temperatura[1]   # componente em x
```

```python
fig, ax = plt.subplots()
ax.quiver(X, Y, -dZ_dx, -dZ_dy)
ax.contour(X, Y, Z)
```
