```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# Dados
x = np.linspace(0, 2*np.pi, 1000, endpoint=False)
y = np.sin(x)

# Cria o slider
valor_alvo = widgets.FloatSlider(
    value=np.pi / 2,
    min=0,
    max=2 * np.pi,
    step=0.01,
    description='valor alvo:',
    continuous_update=True
)
display(valor_alvo)

# Função que atualiza o gráfico
def atualizar_grafico(v):
    clear_output(wait=True)  # limpa o output anterior
    indice = np.argmin(np.abs(x - v))
    x_mais_proximo = x[indice]
    y_mais_proximo = y[indice]

    delta_x = x[indice+1] - x[indice-1]
    delta_y = y[indice+1] - y[indice-1]
    derivada_alvo = delta_y / delta_x

    coef_linear = y_mais_proximo - derivada_alvo * x_mais_proximo
    x_reta = np.linspace(x_mais_proximo-max(x)/10, x_mais_proximo+max(x)/10, 100)
    y_reta = derivada_alvo * x_reta + coef_linear

    # redesenha apenas o gráfico e o slider existente
    display(valor_alvo)
    plt.title("Função e reta tangente")
    plt.xlabel("$x$")
    plt.ylabel("$y$")
    plt.grid()
    plt.plot(x, y, label='y = sin(x)')
    plt.scatter(x_mais_proximo, y_mais_proximo, color='red')
    plt.plot(x_reta, y_reta, color='orange', label='reta tangente')
    plt.legend()
    plt.show()

# Conecta o slider à função
widgets.interactive_output(atualizar_grafico, {'v': valor_alvo})
```
