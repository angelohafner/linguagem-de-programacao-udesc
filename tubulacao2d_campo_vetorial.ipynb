# Definição inicial do campo vetorial

Define-se o campo como $\vec v = 0 \hat i + x(x-d) \hat j$

```python
import numpy as np
import matplotlib.pyplot as plt

diametro = 1
altura = 4
nr_pontos = 11

x = np.linspace(0, diametro, nr_pontos)
y = np.linspace(0, altura, int(nr_pontos/3))

vx = 0 * x
vy = -x*(x-diametro)

plt.quiver(x, altura/2, vx, vy, scale_units='xy', angles='xy', scale=1)
plt.ylim(2,2.5)
```

```python
vy
```

```python
X, Y = np.meshgrid(x, y)
plt.scatter(X,Y)
VX = 0 * X
VY = -X*(X-diametro)
plt.quiver(X, Y, VX, VY, scale_units='xy', angles='xy', scale=0.2)
plt.ylim(-1, 1.3*altura)
plt.grid()
```

Calcula-se o rotacional analiticamente

```python
import sympy as sp

# Symbols
xana, yana, zana, diametro = sp.symbols('x y z diametro')

# Vector field A = (Ax, Ay, Az)
Ax = 0
Ay = xana * (xana - 1)
Az = 0

# Curl A = ∇ x A
compX_rot = sp.diff(Ay, zana) - sp.diff(Az, yana),  # dAy/dz - dAz/dy
compY_rot = sp.diff(Az, xana) - sp.diff(Ax, zana),  # dAz/dx - dAx/dz
compZ_rot = sp.diff(Ax, yana) - sp.diff(Ay, xana)   # dAx/dy - dAy/dx

compZ_rot
```

Reresenta-se o módulo do rotacional.

```python
plt.scatter(X,Y)
VX = 0 * X
VY = -X*(X-1)
cf = plt.contourf(X,Y, 1-2*X, alpha=0.5, cmap="viridis")
plt.colorbar(cf, label="rot")
plt.quiver(X, Y, VX, VY, scale_units='xy', angles='xy', scale=0.2)
plt.ylim(-1, 1.3*altura)
plt.grid()
```

```python
Z = 0*X
VX = 0 * X
VY = -X*(X-1)
VZ = 0 * VX




fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(X, Y, Z)
ax.quiver3D(X, Y, Z, VX, VY, VZ, arrow_length_ratio=0.02, length=5)
ax.quiver3D(X, Y, Z, 0*VX, 0*VY, 1-2*X, arrow_length_ratio=0.2, length=0.05)
plt.show()
```
