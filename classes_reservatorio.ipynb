```python
# Class representing a petroleum reservoir
class Reservoir:
    def __init__(self, name, area, thickness, porosity, water_saturation):
        self.name = name
        self.area = area                # in acres
        self.thickness = thickness      # in feet
        self.porosity = porosity        # fraction (0–1)
        self.water_saturation = water_saturation  # fraction (0–1)

    def ooip(self):
        # OOIP formula: 7758 * A * h * φ * (1 - Sw)
        ooip = 7758 * self.area * self.thickness * self.porosity * (1 - self.water_saturation)
        return ooip

    def show_info(self):
        print(f"Reservoir: {self.name}")
        print(f"  Area = {self.area} acres")
        print(f"  Thickness = {self.thickness} ft")
        print(f"  Porosity = {self.porosity}")
        print(f"  Water Saturation = {self.water_saturation}")
        print(f"  OOIP = {self.ooip():,.2f} bbl")

# Example usage
r1 = Reservoir("R1", area=500, thickness=50, porosity=0.18, water_saturation=0.25)
r2 = Reservoir("R2", area=800, thickness=60, porosity=0.22, water_saturation=0.20)

r1.show_info()
r2.show_info()

# Compare OOIP
if r1.ooip() > r2.ooip():
    print("Reservoir R1 has larger OOIP.")
else:
    print("Reservoir R2 has larger OOIP.")
```
