```python
# Class representing an oil well
class Well:
    def __init__(self, name, depth, pressure, flow_rate):
        self.name = name
        self.depth = depth      # meters
        self.pressure = pressure  # psi
        self.flow_rate = flow_rate  # m3/day

    def productivity_index(self):
        # Simplified PI calculation (J = Q / Δp)
        delta_p = self.pressure - 500  # reference bottomhole pressure
        J = self.flow_rate / delta_p
        return J

    def show_info(self):
        print(f"Well {self.name}: Depth={self.depth}m, Pressure={self.pressure}psi, Flow={self.flow_rate}m³/d")

# Example usage
w1 = Well("P1", 2500, 3000, 120)
w2 = Well("P2", 1500, 2000, 220)

print(w1.show_info())
print(w2.show_info())
```
