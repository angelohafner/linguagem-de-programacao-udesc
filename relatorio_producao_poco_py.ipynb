```python
from typing import List

# Declaracao de variaveis com tipos explicitos
meta_bpd: float = 1200.0

numero_dias: int = 0
indice: int = 0

producao_dia: float = 0.0
total_producao: float = 0.0
media_producao: float = 0.0
maior_producao: float = 0.0
menor_producao: float = 0.0

dias_acima_meta: int = 0
dias_parados: int = 0

producao_diaria: List[float] = []

# Leitura do numero de dias
numero_dias = int(input("Numero de dias para analise: "))

# Loop de entrada de dados com validacao
indice = 0
while indice < numero_dias:
    producao_dia = float(input(f"Producao no dia {indice + 1} (bpd): "))

    if producao_dia < 0.0:
        print("A producao nao pode ser negativa.")
    else:
        producao_diaria.append(producao_dia)
        indice = indice + 1

# Inicializacao dos valores minimo e maximo
maior_producao = producao_diaria[0]
menor_producao = producao_diaria[0]

# Loop de processamento
for producao_dia in producao_diaria:
    total_producao = total_producao + producao_dia

    if producao_dia > maior_producao:
        maior_producao = producao_dia

    if producao_dia < menor_producao:
        menor_producao = producao_dia

    if producao_dia >= meta_bpd:
        dias_acima_meta = dias_acima_meta + 1

    if producao_dia == 0.0:
        dias_parados = dias_parados + 1

# Calculo da media
media_producao = total_producao / numero_dias

# Relatorio de saida
print("\n=== RELATORIO DE PRODUCAO DO POCO ===")
print("Producao diaria (bpd):", producao_diaria)
print("Producao media (bpd):", round(media_producao, 2))
print("Maior producao (bpd):", maior_producao)
print("Menor producao (bpd):", menor_producao)
print("Dias acima da meta:", dias_acima_meta)
print("Dias parados:", dias_parados)
```
